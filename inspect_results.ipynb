```python
# comparison_path = "output/only_spectra/forward_model_only_spectra_model_predicted_data_x2_1.0_5_wpsf.fits"
# comparison_path = "output/only_spectra/forward_model_only_spectra_model_predicted_data_x2_1.0_100_wpsf.fits"
#comparison_path = "output/only_spectra/forward_model_only_spectra_model_predicted_data_x2_5.0_1_wpsf.fits"
#comparison_path = "output/only_spectra/forward_model_only_spectra_model_predicted_data_x2_5.01_1.01_wpsf.fits"
#comparison_path = "output/img_norm/forwardmodel_img_normalized_model_predicted_data_x2_1.0_5_wpsf.fits"
comparison_path = "output/img_norm/forwardmodel_img_normalized_model_predicted_data_x2_1.01_0.1_wpsf.fits"
comparison_path = "output/img_norm/forwardmodel_img_normalized_model_predicted_data_x2_5.0_1_wpsf.fits"
comparison_path = "output/img_norm/forwardmodel_img_normalized_model_predicted_data_x2_9.0_0.1_wpsf.fits"
comparison_path = "output/img_norm/forwardmodel_img_normalized_model_predicted_data_x2_1.0_6_wpsf.fits"
comparison_path = "output/full_data/combined_ECCCO_sw_lw_s_i_scaled_model_predicted_data_x2_1.0_0.2_wpsf.fits"
#comparison_path = "output/img_norm/forwardmodel_img_normalized_model_predicted_data_x2_10.0_6.02_wpsf.fits"
comparison = fits.getdata(comparison_path)

ground_truth_path = "data/forwardmodel_img_normalized.fits" # "data/eccco_is_lw_forwardmodel_thermal_response_psf4pix_el.fits"
ground_truth_path = "data/ECCCO_speedtest_runs/combined_ECCCO_sw_lw_s_i_scaled.fits"

ground_truth = fits.getdata(ground_truth_path)
```

```python
vmin, vmax = 0, 25

fig, axs = plt.subplots(nrows=2, figsize=(20, 5), sharex=True)
axs[0].imshow(np.sqrt(np.abs(ground_truth)) * np.sign(ground_truth), origin='lower', vmin=vmin, vmax=vmax)
axs[1].imshow(np.sqrt(comparison), origin='lower', vmin=vmin, vmax=vmax)
fig.savefig("./output/full_data/prediction.png")
```

```python
vmin, vmax = -100, 100

fig, ax = plt.subplots(figsize=(20, 5), sharex=True)
im = ax.imshow(comparison - ground_truth, origin='lower', vmin=vmin, vmax=vmax, cmap='seismic')
ax.set_title("prediction - input")
fig.colorbar(im, ax=ax)
```
